Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(
    np.array(data['LSTAT']).reshape(-1,1),
    np.array(data['MEDV']),
    test_size = 0.2,
    random_state = 1000
)
```

```python
model.fit(X_train, y_train)
predicted_y = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title('Linear Regression')


x = np.array([0, 50])
y = model.coef_[0]*x + model.intercept_
plt.plot(x,y)
plt.plot(np.array(data['LSTAT']).reshape(-1,1), 
         np.array(data['MEDV']), 
         'r.', alpha = 0.2)
```

**Регрессионный анализ** - можно ли провести линейную зависимость между двумя выборками. Возьмем Ladder - удовлетворение жизнью, и generosity - доброта. Мы хотим найти прямую, которая разделит точки в пространстве наполовину. Это можно сделать графически. Нарисовать график. Можно ли сказать, что данная прямая является собственным вектором? Сделать график регрессионной прямой на данных и собственных векторов. 

```python
import numpy.linalg as la
import math
from sklearn.preprocessing import StandardScaler

data_with_cont = data_with_cont.dropna()
lad = data_with_cont['Ladder']
gen = data_with_cont['Generosity']

Y = np.vstack((lad, gen)).T
w,v = la.eig(np.cov(Y.T))

sns.regplot(
    x=lad, 
    y=gen)
for i in range(len(w)):
    fig = plt.gcf()
    fig.set_size_inches(6,6)
    print('Собственное значение: ', round(w[i],2), 'Собственный вектор: ', v[:,i])
    plt.xlabel('Ladder')
    plt.ylabel('Generosity')
    plt.plot([lad.mean(), lad.mean() + math.sqrt(w[i]) * v[:,i][0]], 
             [gen.mean(), gen.mean() + math.sqrt(w[i]) * v[:,i][1]], color='red', linewidth=3)
    
```

В данных почти не наблюдается линейной зависимости.
Регресионная прямая не является собственным вектором (тк она не совпадает ни с одним собственным вектором).

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
data_uni = stats.uniform.rvs(loc=1, scale=6, size=10, discrete='discrete')
print(data_uni)

ax = sns.distplot(data_uni,
                 color="skyblue")
ax.set(xlabel='Значение', ylabel='Вероятность')
plt.title('Плотность рапределения Равномерного')
```
